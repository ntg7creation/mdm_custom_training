```python
# view_dmvb.py
import numpy as np

PATH = "../converted_motions/hand_poses/p001-folder/keypoints_3d/000/dmvb_left.npy"
motion = np.load(PATH)

print("DMVB shape:", motion.shape)

# Index breakdown
ROOT_ROT_Y     = 0
ROOT_VEL_XZ    = slice(1, 3)
ROOT_Y         = 3
RIC            = slice(4, 4 + 63)
ROT6D          = slice(4 + 63, 4 + 63 + 126)
LOCAL_VEL      = slice(4 + 63 + 126, 4 + 63 + 126 + 66)
CONTACT        = slice(4 + 63 + 126 + 66, 263)

# Preview slices
print("Root Y-rotation velocity:\n", motion[:, ROOT_ROT_Y])
print("Root linear velocity (XZ):\n", motion[:, ROOT_VEL_XZ])
print("Root height (Y):\n", motion[:, ROOT_Y])
print("RIC shape:", motion[:, RIC].shape)
print("ROT6D shape:", motion[:, ROT6D].shape)
print("Local velocity shape:", motion[:, LOCAL_VEL].shape)
print("Foot contact shape:", motion[:, CONTACT].shape)

# Check for NaNs or static frames
print("Contains NaNs:", np.isnan(motion).any())
print("All-zero frames:", np.any(np.all(motion == 0, axis=1)))
```
